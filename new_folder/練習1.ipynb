#### 0.3. ループ

ループについては以下のように記述します

イテラブルオブジェクトはリストや辞書、文字列などである

```py
for 変数名 in イテラブルオブジェクト:
    処理
```

````python
for i in range(1,5):
    print(i)
# 1
# 2
# 3
# 4
````

for文が正常に終了した後に処理を行う場合は`else`を使います

````python
for i in range(1,5):
    print(i)
else:
    print("ループが終了")
````

`enumerate()`でオブジェクトとインデックスを同時に取得できます

````python
l = ["A","B","C","D"]
for i, name in enumerate(l):
    print(i, name)
````

#### 0.5. クラス

````python
class Complex:
    def __init__(self, realpart, imagpart): # インスタンスを作成したときの初期化内容を記述
        self.r = realpart # self.変数名とする
        self.i = imagpart

x = Complex(3.0, -4.5)
print(x.r, x.i)
````

インスタンスオブジェクトに属性を追加することができます

````python
x = Complex(3.0, -4.5)
x.h = 10
print(x.__dict__)
# {'r': 3.0, 'i': -4.5, 'h': 10}
````

### 1. 関数

関数は以下のように定義します
```py
def 関数名(引数1,引数2,…);
    処理
    return 戻り値
```

以下のように関数を呼びだします
```py
関数名(引数1,引数2,…)
```

以下のように`c=100`とデフォルト引数を設定できます。デフォルト引数は引数として何も指定されなかったときに使用されます

````python
def func_default(a, b, c=100):
    print(f'a={a}, b={b}, c={c}')

func_default(1, 10)
func_default(1, 10, 200)
````

````python
def func_default(a=1, b, c):
    print(f'a={a}, b={b}, c={c}')
````

このようにデフォルト引数を指定した後に省略することはできません

デフォルト引数値は関数定義が実行されるときに評価されるので、デフォルト引数に更新可能なオブジェクトを指定すると予期せぬ動作をすることがあります : [[参考]](https://note.nkmk.me/python-argument-default/)

`*args`のように`*`を変数につけると複数の引数をタプルとして受け渡すことができます

````python
def func_args(*args):
    print(args)

func_args(1, 10)
func_args(1, 10, 100, 1000)
````

`**kwargs`のように`**`を変数につけると複数のキーワード引数が辞書として渡されます

````python
def func_kwargs(**kw):
    name = "Taro"
    if "name" in kw:
        name = kw["name"]
    print(name)

kw = {}
func_kwargs(**kw)
kw = {"name" : "Jiro"}
func_kwargs(**kw)
````

呼び出し時にリストやタプルに`*`をつけて指定すると、要素が展開（アンパック）され順番に位置引数として指定されます

````python
def func(a, b, c):
    print(f'a={a}, b={b}, c={c}')

l = [1, 10, 100]
func(*l)
````

関数呼び出し時に辞書に`**`をつけて指定すると、要素のキーが引数名、値が引数の値として展開されてキーワード引数として指定されます

````python
d = {'a': 1, 'b': 10, 'c': 100}
func(**d)
````

返り値(return)について、複数の値をカンマ区切りで指定するとタプルが返されます

````python
def func_return_multi(a, b):
    return a + b, a * b, a / b

x, y, z = func_return_multi(3, 4)
print(x)
l = func_return_multi(3, 4)
print(l)
````

引数に`*`を含めると、それ以降の引数はキーワード専用引数となります

````python
def func(a, * , b, c):
    print(a,b,c)
````

````python
func(1,2,3) # キーワード専用のため、エラーとなる
````

````python
func(1,b=2,c=3)
````
